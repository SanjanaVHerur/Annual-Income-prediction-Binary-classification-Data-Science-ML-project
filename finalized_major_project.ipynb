```python
x = df.drop(['income'], axis = 1)
y = df['income']
'''
# Verification
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
x.head()
y.head()
'''
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

### Train and test data splitting

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Confusion Matrix and Classification report function

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def gen_metrics(ytest,ypred):
    cm = confusion_matrix(ytest,ypred)
    print('Confusion Matrix :-\n',cm)
    print('\nClassification Report :-\n',classification_report(ytest,ypred))
    print('\nAccuracy Score : ',accuracy_score(ytest,ypred))

def train_test_score(m):
    print('\nTraining Score :', m.score(x_train,y_train))
    print('Testing Score/Accuracy Score :',m.score(x_test,y_test),'\n')
```

### a) DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier

m1 = DecisionTreeClassifier(criterion = 'gini', max_depth = 8,min_samples_split = 20)
m1.fit(x_train,y_train)

ypred_m1 = m1.predict(x_test)
df1 = pd.DataFrame({'y_test':y_test,'ypred_m1':ypred_m1})
df1.head()
#Predicted Values
```

```python
gen_metrics(y_test,ypred_m1)
train_test_score(m1)
```

Validating the result

```python
#Precision

precision_positive = precision_score(y_test, ypred_m1, pos_label=1)
precision_negative = precision_score(y_test, ypred_m1, pos_label=0)
print(precision_positive, precision_negative) 
```

```python
#Recall

recall_positive = recall_score(y_test, ypred_m1, pos_label=1)
recall_negative = recall_score(y_test, ypred_m1, pos_label=0)
print(recall_positive, recall_negative) 
```

```python
#F1 score

f1_score_positive = f1_score(y_test, ypred_m1, pos_label=1)
f1_score_negative = f1_score(y_test, ypred_m1, pos_label=0)
print(f1_score_positive, f1_score_negative) 
```

```python
#Accuracy

print('Accuracy: %.3f' % accuracy_score(y_test, ypred_m1))
```

```python
#misclassification percentage

misclassification_rate = 1-(accuracy_score(y_test, ypred_m1))
print(misclassification_rate)
misclassification_percentage = misclassification_rate*100
print('Misclassification_percentage = ',misclassification_percentage)
```

#### The precision,recall,f1-score and accuracy values present in classification report are same as when calculate using confusion matrix and Hence verified.
